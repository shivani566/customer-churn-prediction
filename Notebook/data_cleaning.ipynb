```python
import pandas as pd
import numpy as np

# Load raw data
df = pd.read_csv(r"C:\Users\vlogs\OneDrive\Desktop\Churn Prediction\Data\Telco-Customer-Churn.csv")

# Fix 'TotalCharges' column
df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan).astype(float)
df['TotalCharges'].fillna(df['MonthlyCharges'] * df['tenure'], inplace=True)

# Check SeniorCitizen type
print(df['SeniorCitizen'].dtype)  # Should be int or float

# Leave SeniorCitizen as is â€” it's numeric (0 or 1). Do NOT map or convert unless necessary

# Strip whitespace from string columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()

# Fill any other NaNs if found (check first)
print(df.isna().sum())

# If any categorical columns have NaNs, fill with mode:
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

# For numerical columns (other than TotalCharges handled above), fill NaN with median or mode
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    if df[col].isna().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)

# Save cleaned data for further processing if you want:
df.to_csv("cleaned_churn_data.csv", index=False)

#No null values detected
```
