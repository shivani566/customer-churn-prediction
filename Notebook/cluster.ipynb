```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load cleaned data
df = pd.read_csv(r'C:\Users\vlogs\OneDrive\Desktop\Churn Prediction\Data\cleaned_churn_data.csv')

# Select numeric columns for clustering
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

# Handle missing values if any
df = df.dropna(subset=num_cols)

# Scale numeric features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[num_cols])

# Apply KMeans clustering
k = 4  # You can experiment with this
kmeans = KMeans(n_clusters=k, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Optional: map cluster numbers to labels (Segment 1, 2, 3, 4)
df['Segment'] = df['Cluster'].apply(lambda x: f'Segment {x+1}')

# Save to CSV for Power BI
df.to_csv(r'C:\Users\vlogs\OneDrive\Desktop\Churn Prediction\Notebook\churn_with_clusters.csv', index=False)
```
